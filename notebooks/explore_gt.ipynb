```python
from pathlib import Path

import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv(Path("../../data/output.csv"))
df.head()
```

```python
# Plot the distribution of the ground truth

sns.scatterplot(data=df, x="PHQ9", y="BDI")
```

```python
# Transform data into classes
bdi_classes = pd.cut(df["BDI"], bins=[0, 9, 18, 29, 63], labels=["minimal", "mild", "moderate", "severe"])
phq9_classes = pd.cut(df["PHQ9"], bins=[0, 4, 9, 14, 27], labels=["minimal", "mild", "moderate", "severe"])

# Plot the distribution of the classes
g = sns.histplot(
    x=bdi_classes[::-1],
    y=phq9_classes[::-1],
)
g.invert_yaxis()
```
